# Cosa abbiamo scoperto — da raccontare a un amico

Ti ricordi le domande di quelle sere a commentare i giornali? *"Ma Francesco è
comunista? Ha rotto coi Papi di prima? Parla solo di migranti?"*. Domande vere,
sul senso delle cose e sulla fede. A un certo punto ci siamo detti: invece di
tirare a indovinare, **guardiamo i dati**.

Abbiamo raccolto **~25.000 testi** dei quattro Papi (Giovanni Paolo II, Benedetto
XVI, Francesco, Leone XIV), li abbiamo indicizzati e interrogati. Qui sotto, per
ogni domanda: l'**ipotesi** da cui partivamo, la **domanda** secca, l'**analisi**
che abbiamo fatto e la **conclusione** che dicono i numeri. Tutto su viste
aggregate — niente testi ripubblicati.

> Le analisi per esteso sono nei notebook `01`–`05` di questa cartella; qui è il
> racconto.

## 1. Di cosa parlano *davvero*?

**Ipotesi.** Dai giornali sembra che ogni Papa "sia" un tema: Francesco i
migranti, Wojtyła i viaggi… come se parlassero soprattutto di quello.

**Domanda.** Se prendo tutto quello che dice un Papa, di cosa è fatto?

**Analisi.** Abbiamo strutturato gli argomenti in **sei famiglie** e marcato ogni
*pezzo* di discorso con la sua (i discorsi si mischiano: un'omelia può avere
liturgia *e* attualità). → notebook `04`.

**Conclusione.** Di base parlano tutti della **stessa cosa**: la **linea rossa**
liturgico-devozionale (Dio, Gesù, Maria, i sacramenti, il Vangelo) è circa
l'**80%** di tutto, per ognuno dei quattro. I temi "da giornale" vivono nel **20%
restante**: lì sì che si vedono gli accenti — l'**attualità** per Francesco e
Leone, i **viaggi** per Giovanni Paolo II.

![struttura per famiglie](sintesi-struttura.png)

## 2. C'è continuità tra i Papi?

**Ipotesi.** "Con Francesco è cambiato tutto, ha rotto con la tradizione."

**Domanda.** I temi portanti sono gli stessi da un pontificato all'altro, o no?

**Analisi.** Conteggio dei temi per Papa, sia a **parole chiave** sia per
**significato** (gli embedding), con un indice `lift` che dice chi sta sopra o
sotto la media. → notebook `01`.

**Conclusione.** La **continuità è schiacciante.** I temi che dominano sono
identici per tutti e quattro: Dio/Vangelo (~85%), pace (~50%), famiglia (~52%) —
con `lift` attorno a 1 per ognuno. Su questo i Papi sono **la stessa voce**.
Francesco non è un corpo estraneo: cambia gli accenti, non la sostanza.

![continuità](sintesi-continuita.png)

## 3. Francesco è comunista?

**Ipotesi.** "Parla sempre di poveri, migranti, disuguaglianze: è di sinistra,
quasi comunista."

**Domanda.** Davvero questi temi sono *suoi*? E parlare di poveri e operai fa di
un Papa un comunista?

**Analisi.** Stessi conteggi del punto 2, concentrati sui temi "sociali" (poveri,
migranti, giustizia, lavoro), confrontati tra i quattro. → notebook `01`.

**Conclusione.** Sì, Francesco **accentua** davvero poveri, migranti e
disuguaglianze (`lift` 1,2–1,5): è il suo timbro. Ma sono temi **minori** rispetto
a Dio/pace/famiglia, e **li trattano tutti** — è la dottrina sociale della Chiesa,
vecchia di oltre un secolo, non comunismo. La prova del nove è la riga
**lavoro/operai**: il primo è **Giovanni Paolo II** (il doppio di Francesco) —
proprio il Papa che ha fatto *cadere* il comunismo. Parlare di poveri non rende
comunisti.

![comunista?](sintesi-comunista.png)

> Stessa storia per l'**ambiente**: quasi uguale per tutti (~16–19%). Non l'ha
> inventato Francesco; di suo c'è solo il *modo di dirlo*, la formula "casa comune"
> della *Laudato si'*. (Il quadro completo dei temi, parola vs significato, è nella
> heatmap del notebook `01`.)

![heatmap dei temi](temi_per_papa.png)

## 4. Hanno tenuto quello che avevano promesso?

**Ipotesi.** "Dicono una cosa appena eletti e poi se ne dimenticano."

**Domanda.** Il programma dichiarato all'inizio (l'omelia di inizio pontificato)
regge per tutto il pontificato, o si perde per strada?

**Analisi.** Abbiamo preso il **baricentro semantico** dell'omelia inaugurale come
"vettore-mandato" e misurato, **anno per anno**, quanto i documenti gli restano
vicini. → notebook `05`.

**Conclusione.** **Tutti tengono il mandato.** La fedeltà resta **piatta** per
l'intero pontificato — Giovanni Paolo II su **27 anni** non si allontana. Nessuna
deriva: la linea dichiarata all'inizio è quella di sempre.

![tenuta del mandato](sintesi-mandato.png)

> Da leggere la **piattezza di ogni linea**, non quanto sta in alto: ogni Papa è
> confrontato col *suo* mandato, quindi i livelli non sono paragonabili tra loro
> (e le similarità del modello stanno comunque tutte alte). Il segnale è
> l'assenza di pendenza.

## La morale

I dati smontano il titolo di giornale e fanno vedere la **continuità**. Non "un
Papa contro gli altri", ma una voce che cambia accento e parole su un filo che
resta. In una riga:

> **Francesco cambia gli accenti, non la sostanza. Continuità piena, comunismo no.
> E ognuno, alla sua maniera, tiene la linea dichiarata il primo giorno.**
